# Get Films URLs

```python
from allocine_scraper import get_film_urls

urls = get_film_urls(ROOT_URL)
```

## Store to file

```python
# Warning: this will overwrite the url file
with open(URL_FILE, 'w') as f:
    for url in urls:
        f.write("{}\n".format(url))
```

## Read from file

```python
urls = []
with open(URL_FILE, 'r') as f:
    urls = f.read().splitlines()
print(len(urls))
```

# Scrap Films from URLs

```python
batches = [
    # The first films are very popular
    # urls[0]
    urls[0:1000], # 0
    urls[1000:2000], # 1
    urls[2000:3000], # 2
    urls[3000:4000], # 3
    urls[4000:5000], # 4
    urls[5000:6000], # 5
    urls[6000:7000], # 6
    urls[7000:8000], # 7
    urls[8000:9000], # 8
    urls[9000:10000], # 9 
    # urls[10000:20000], # 10
    # urls[20000:30000], # 11
    # urls[30000:40000], # 12
    # urls[40000:50000], # 13
    # urls[50000:60000], # 14
    # urls[60000:70000], # 15
    # urls[70000:80000], # 16
    # urls[80000:90000], # 17
    # urls[90000:100000], # 18
    # urls[100000:110000], # 19
    # urls[110000:], # 20
]
```

```python
import os
import pandas as pd
from allocine_scraper import get_film_reviews
import re

OUTPUT_PATH = "./pickle"
MAX_REVIEWS_PER_FILM = 30

def makedir(path):
    if not os.path.exists(path):
        os.makedirs(path)

makedir(OUTPUT_PATH)

for i, batch_urls in enumerate(batches):    
    print("Batch {}/{}".format(i+1, len(batches)))
    #if i < 19:
    #    continue 
    if(i < 4):
        continue  
    
    # Determine pickle file name
    out_file = "allocine_{}.pickle".format(i)
    out_path = os.path.join(OUTPUT_PATH, out_file)
    print(f'out path is: {out_path}')
    
    dic = get_film_reviews(ROOT_URL, batch_urls, MAX_REVIEWS_PER_FILM)
    print()
    
    # Create a pandas DataFrame and save it to disk
    df = pd.DataFrame.from_dict(dic)
    df.to_pickle(out_path)


####################################################################################################

# out_file = "allocine_00.pickle"
# out_path = os.path.join(OUTPUT_PATH, out_file)

# film_id = re.findall(r'\d+', urls[0])[0]
# film_url = "{root}/film/fichefilm-{film_id}/critiques/spectateurs/".format(
#             root=ROOT_URL,
#             film_id=film_id
#         )
# print(film_url)
# # # parsed = parse_film(film_url, 30)
# parsed = get_film_reviews(ROOT_URL, urls[0:100], MAX_REVIEWS_PER_FILM)

# # # dic = get_film_reviews(ROOT_URL, urls[0:10], MAX_REVIEWS_PER_FILM)
# df = pd.DataFrame.from_dict(parsed)
# print(df)
# # for i, batch_urls in enumerate(batches):    
# #     print("Batch {}/{}".format(i+1, len(batches)))
# #     #if i < 19:
# #     #    continue        
# #     # Determine pickle file name
# #     out_file = "allocine_{}.pickle".format(i)
# #     out_path = os.path.join(OUTPUT_PATH, out_file)
# #     print(f'out path is: {out_path}')
    
# #     dic = get_film_reviews(ROOT_URL, batch_urls, MAX_REVIEWS_PER_FILM)
# #     print()
    
# #     # Create a pandas DataFrame and save it to disk
# #     df = pd.DataFrame.from_dict(dic)
# #     df.to_pickle(out_path)
```
